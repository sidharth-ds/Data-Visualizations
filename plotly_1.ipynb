# Line graph:

```python
x1 = [1,2,3,4,5]
y1 = [1,23,3,4,5]

fig = px.line(x=x1,y=y1,title='sample')
fig.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [1,23,3,4,5]
y2 = [11,23,31,4,5]

fig = px.line(x=x1,y=y1,title='sample-y1') 
fig = px.line(x=x1,y=y2,title='sample-y2')  # 2nd graph will replace 1st graph
fig.show()
```

# Bar graph:

```python
import numpy as np
```

```python
x_=np.random.randint(1,100,100)
y_=np.random.randint(1,100,100)

fig = px.bar(x=x_,y=y_)
fig.show()
```

```python
for i in range(len(x_)):   # value of y get added, if x values are same.
  print(x_[i],y_[i])
```

```python
fig = px.bar(x=x_,y=y_,orientation='h')  # barh
fig.show()
```
