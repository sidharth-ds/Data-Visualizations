```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

# defining x & y axis:

t = x-axis

y1,y2 = y-axis:

```python
t = np.arange(1,11)
t
```

```python
y1 = [4,1,7,-8,2,9,5,8,3,6]
y2 = [-6,-3,8,6,-1,1,2,3,9,-7]
```

x-axis should be sorted in ascending order for better visualization.

# line graph:

```python
plt.plot(t,y1)
plt.show()
```

if x-axis is not given, it takes the index values of y-axis [0,1,2,...] as x-axis.

x-axis starts from zero.

```python
plt.plot(y1)    
plt.show()          
```

```python
plt.plot(t,y2)    
plt.show()
```

2nd graph will be over and above the 1st graph:

```python
plt.plot(t,y1)
plt.plot(t,y2)
plt.show()        
```

label : is the name of a graph.

legend() : is used to call the label:

```python
plt.plot(t,y1,label = 'car')
plt.plot(t,y2,label = 'bike')
plt.legend()
plt.show()
```

xlabel, ylabel : is the name of the axis:

title : is the name of the figure:

```python
plt.plot(t,y1,label="car")
plt.plot(t,y2,label="bike")

plt.xlabel("distance")
plt.ylabel("speed")
plt.title("sample graph")

plt.legend()
plt.show()
```

for marker: https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(t,y1,marker = 'o')   
plt.show()
```

ms - marker size

```python
plt.plot(t,y1, marker = '*', ms=10)  
plt.show()
```

for color: https://matplotlib.org/stable/gallery/color/named_colors.html

```python
plt.plot(t,y1,marker='*',ms=15,mec='k')     
plt.show()
```

mfc - Marker Face Colour

mec - Marker Edge Colour

```python
plt.plot(t,y1,marker='*',ms=15,mec='k',mfc='y')   
plt.show()                                        
```

for linestyle:
https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
plt.plot(t,y1,linestyle='dashdot',color='m')      
plt.show()
```

color, linewidth:

```python
plt.plot(t,y1,color='c',linewidth='8')
plt.plot(t,y2,color='r',linewidth='6')
plt.show()
```

# bar graph:

```python
t = np.arange(1,11)
t,y1,y2
```

```python
plt.bar(t,y1)
plt.show()
```

```python
plt.bar(t,y2)
plt.show()
```

y2 graph will overlap y1 graph:

```python
plt.bar(t,y1,color='k')     
plt.bar(t,y2,color='y')
plt.show()
```

y1 graph will overlap y2 graph:

```python
plt.bar(t,y2,color='y')   
plt.bar(t,y1,color='k')
plt.show()
```

width:

```python
plt.bar(t,y1,width=0.5)
plt.show()
```

```python
plt.bar(t,y1,width=0.8)
plt.show()
```

in case of barh, height is given:

```python
plt.barh(t,y2,height=0.5)
plt.show()
```

```python
plt.barh(t,y2,height=0.8)
plt.show()
```

# multiple graphs in same figure:

```python
plt.plot(t,y1,color='r',label='100')
plt.legend()
plt.show()

plt.bar(t,y1,color='g')
plt.title('sample bar')
plt.show()

plt.plot(t,y1,linestyle='dotted',color='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# multiple graphs for same values:

```python
y= np.random.randint(50,100,size=100)
y
```

```python
x = [i for i in range(1,101)]
```

```python
plt.plot(x,y)
plt.title('plot')
plt.show()
```

```python
plt.bar(x,y)
plt.title('bar')
plt.show()
```

# histogram:

```python
x = np.random.randint(50,100,size=100)
x
```

```python
plt.hist(x)   # y-axis is automatic, since it is a count
plt.show()
```

```python
plt.hist(x,bins=2)
plt.show()
```

```python
plt.hist(x,bins=10)
plt.show()
```

```python
plt.hist(x,bins=20)
plt.show()
```

cumulative: adds the previous values.

```python
plt.hist(x,cumulative=True,bins=10)
plt.show()
```

calculations for bins:

```python
x.min(),x.max()
```

```python
x.max() - x.min()
```

```python
(x.max()-x.min()) / 10      # each bin length 
```

```python
x.min() , x.min() + (x.max()-x.min()) / 10
```

```python
x.min() + (x.max()-x.min()) / 10 ,   x.min() + (((x.max()-x.min()) / 10) * 2 )
```
