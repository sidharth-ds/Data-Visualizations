# importing default Datasets:

```python
import seaborn as sns
```

```python
dir(sns)
```

```python
sns.get_dataset_names()
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

# Hue

basic scatter plot:

```python
sns.scatterplot(x='total_bill',y='tip',data=df1)
plt.show()
```

scatter plot with hue: (without order)

```python
sns.scatterplot(x='total_bill',y='tip',data=df1,hue='smoker')
plt.show()
```

scatter plot with hue: (with order)

```python
sns.scatterplot(x='total_bill',y='tip',data=df1,hue='smoker',hue_order=['No','Yes'])
plt.show()
```

changing colors:

```python
hue_colors={'Yes':'black','No':'red'}
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df1,hue='smoker', palette=hue_colors)
plt.show()            
```

using hue with countplot:

```python
sns.countplot(x='smoker',data=df1,hue='sex')
plt.show()
```

# realtional plots and subplots:

to establish relationship between variables.

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='sex',kind='scatter')
plt.show()
```

subplot in relational plot:

1. column-wise
2. row-wise

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='sex',kind='scatter',col='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='sex',kind='scatter',row='smoker')
plt.show()
```

subplots in rows and column:

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='sex',kind='scatter',row='time',col='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='sex',kind='scatter',row='smoker',col='day')
plt.show()
```

# wrapping and ordering of columns:

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='sex',kind='scatter',col='day',col_wrap=4,col_order=['Sat','Sun','Thur','Fri'])
plt.show()
```

# point size:

```python
df1.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,kind='scatter',hue='size',size='size')
plt.show()
```

# changing point transparency:

alpha : between 0 & 1

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='size',kind='scatter',size='size',alpha=0.8)
plt.show()
```
