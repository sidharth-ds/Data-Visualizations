```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = sns.load_dataset('tips')
df2.head()
```

# Line plot:

```python
sns.relplot(x='total_bill',y='tip',data=df2,kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df2,kind='line',style='day',hue='day', markers=True,dashes=False)
plt.show()
```

# Multiple observations per x-value:

```python
sns.relplot(x='day',y='total_bill',data=df2,kind='scatter')
plt.show()
```

```python
sns.relplot(x='day',y='total_bill',data=df2,kind='line')
plt.show()
```

shaded region : indicates confidence interval.(based on mean value)

to turn off: ci=None 

for std deviation: ci='sd' 

```python
sns.relplot(x='day',y='total_bill',data=df2,kind='line',ci=None)
plt.show()
```

# categorical plot: catplot()

easily create subplots with col= and row=

using countplot():

```python
sns.countplot(x='smoker',data=df2)
plt.show()
```

using catplot():

```python
sns.catplot(x='smoker',data=df2,kind='count')
plt.show()
```

changing the order:

```python
category_order = ['No','Yes']
```

```python
sns.catplot(x='smoker',data=df2,kind='count',order=category_order)
plt.show()
```

# Bar plot:

```python
sns.catplot(x='day',y='total_bill',data=df2,kind='bar')
plt.show()
```

to remove the uncertainity indication:

```python
sns.catplot(x='day',y='total_bill',data=df2,kind='bar',ci=None)
plt.show()
```

changing the orientation:

```python
sns.catplot(x='total_bill',y='day',data=df2,kind='bar',ci=None)
plt.show()
```

# Box plot:

to see Median, Spread, Skewness(distribution of the data), Outliers(exceptional data)

application: in time_series data - to analyse the seasonality.

```python
sns.catplot(x='time',y='total_bill',data=df2,kind='box')
plt.show()
```

to change the order:

```python
sns.catplot(x='time',y='total_bill',data=df2,kind='box',order=['Dinner','Lunch'])
plt.show()
```

to omit the outliers:

```python
sns.catplot(x='time',y='total_bill',data=df2,kind='box',sym='')
plt.show()
```

changing the Whiskers using 'whis':

```python
sns.catplot(x='time',y='total_bill',data=df2,kind='box',whis=2.0)
plt.show()
```

show min and max whis values:

```python
sns.catplot(x='time',y='total_bill',data=df2,kind='box',whis=[55,95])
plt.show()
```

# point plot:

```python
sns.catplot(x='smoker',y='total_bill',data=df2,kind='point')
plt.show()
```

to remove the categorical indicator:

```python
sns.catplot(x='smoker',y='total_bill',data=df2,kind='point',join=False)
plt.show()
```

displaying the Median:

```python
from numpy import median
```

```python
sns.catplot(x='smoker',y='total_bill',data=df2,kind='point',estimator=median)
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df2,kind='point',capsize=0.2)
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df2,kind='point',ci=None)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.catplot(x='smoker',y='total_bill',data=df2,kind='point')
plt.show()
```

# changing the color palette:



Diverging palettes: 'RdBu'....'PRGn'....'RdBu_r'....'PRGn_r'  

```python
sns.set_palette('RdBu')
sns.catplot(x='smoker',data=df2,kind='count')
plt.show()
```

Sequential palettes:
* 'Greys'....'Blues'....'PuRd'....'GnBu'  

* it is generally used for scatter plot.

Custom palettes:
* 'red','green','orange','blue','yellow','purple'
* sns.set_palette(custom_plette)


changing the scale:
* 'paper'<'notebook'<'talk'<'poster'
* sns.set_context

```python
sns.set_context('paper')
sns.catplot(x='time',y='total_bill',data=df2,kind='point')
plt.show()

sns.set_context('notebook')
sns.catplot(x='time',y='total_bill',data=df2,kind='point')
plt.show()

sns.set_context('talk')
sns.catplot(x='time',y='total_bill',data=df2,kind='point')
plt.show()

sns.set_context('poster')
sns.catplot(x='time',y='total_bill',data=df2,kind='point')
plt.show()

```
